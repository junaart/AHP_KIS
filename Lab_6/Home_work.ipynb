## Вариант 1.

1.   Пусть $\zeta_1, \zeta_2$  две независимые случайные величины, которые соответственно имеют равномерное и показательное распределения: 
$$f_{\zeta_1}(\zeta_1)=\begin{cases} 1, \zeta_1 \in (0,1)\\
                                     0, \zeta_1 \not\in (0,1)
\end{cases}$$
$$f_{\zeta_2}(\zeta_2)=\begin{cases} e^{-\zeta_2}, \zeta_2\geq 0\\
                                     0, \zeta_1 <0
\end{cases}$$ 
Для случайной величины $Z=\zeta_1+\zeta_2$ определить: $f(Z), M(Z), D(Z)$ (использовать преобразование Лапласа-Стилтьеса).

Все потоки событий —  простейшие. 1) Определить характеристики СМО: среднее число  занятых рабочих; абсолютную пропускную способность; среднее число неисправных станков; 2) Установить, улучшатся ли  характеристики СМО, если рабочие будут налаживать станки совместно, тратя вдвоем на наладку одного станка в среднем 5 мин.

3. Рассматривается простейшая двухканальная СМО с  «нетерпеливыми» заявками. Интенсивность потока заявок $\lambda= 3$ заявки/ч; среднее время обслуживания одной заявки $t_{\text{o6cл}} = 1$ ч; средний срок, в течение которого заявка  «терпеливо» стоит в очереди, равен 0,5 ч. Подсчитать финальные  вероятности состояний. Найти характеристики эффективности СМО: $L_{\text{сист}}, \lambda', T_{\text{сист}}, W_{\text{оч}}, L_{\text{оч}}$.

4. Имеется двухканальная СМО с отказами, на которую поступают два простейших потока заявок: I с интенсивностью $\lambda_1$ и II с  интенсивностью $\lambda_2$. Заявки I имеют перед заявками II приоритет, состоящий в следующем: если заявка I приходит в момент, когда все каналы  заняты и хотя бы один из них обслуживает заявку II, то пришедшая заявка I «вытесняет» из обслуживания заявку II, становится на ее место, а та покидает СМО необслуженной. Если заявка I  приходит в момент, когда все каналы заняты обслуживанием заявок I, то она получает отказ и покидает СМО. Заявка II получает отказ, если она приходит в момент, когда заняты оба канала  (безразлично какими заявками). Построить размеченный граф состояний СМО, нумеруя   состояния двумя индексами $(r,j)$; первый указывает число заявок I, второй — число заявок II, находящихся в СМО. Написать  уравнения для финальных вероятностей состояний. Решить их при $\lambda_1=0.6$, $\lambda_2=1.1,$ $\mu_1=0.3,$ $\mu_2=2$. Найти следующие характеристики эффективности СМО: вероятность отказа в момент поступления заявки I, II; среднее число каналов, обслуживающих заявки I, II; среднее число заявок I (II), обслуживаемое СМО в единицу времени; те же характеристики для СМО в целом,  безотносительно к виду заявок.
