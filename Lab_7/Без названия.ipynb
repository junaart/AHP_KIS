```r
Variant<-1
set.seed(Variant)
m<-sample(c(6:20),1)
lambda<-runif(1,0.1, 2)
lambda = lambda - 0.7 #  лямбда меньше мю
Q<-rexp(m,0.3)

q<-sample(c(1:4),1)
print(data.frame(m,q, lambda))
print(Q)
#среднее время обслуживания
mu <- 1/(sum(Q)/length(Q))
```

```r
library(simmer)
library(simmer.plot)

#reset(mm_2_inf)

my.trajectory <- trajectory() %>%
  seize("server", amount=1) %>%
  timeout(function() rexp(1, mu)) %>%
  release("server", amount=1)

Tmodel<-100000
mm_2_inf <- simmer() %>%
  add_resource("server", capacity=2, queue_size=Inf) %>%
  add_generator("arrival", my.trajectory, function() rexp(1, lambda)) %>%
  run(until=Tmodel)

res_arrivals <- get_mon_arrivals(mm_2_inf)
cat("Woch =", mean(res_arrivals$end_time - res_arrivals$start_time-res_arrivals$activity_time))
View(res_arrivals)
```
